```python
chunk_size = 100000

files = ['2006.csv', '2007.csv']

for file in files:
    with pd.read_csv(file, chunksize = chunk_size) as reader:
        for chunk in reader:
            chunk.to_sql('airlines', con=conn, if_exists='append', index=False
                         , chunksize = chunk_size)
```

```python
cnt = conn.execute("SELECT COUNT(*) FROM airlines")
print(f'There are {[row[0] for row in cnt]} rows in DB')
```
